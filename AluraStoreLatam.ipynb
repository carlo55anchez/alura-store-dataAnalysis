### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto').size()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size()
```

```python
print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_tienda1)
print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_tienda2)
print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_tienda3)
print("\nVentas por categoría - Tienda 4:")
print(ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
```

```python
print("Calificación promedio Tienda 1:", calificacion_promedio_tienda1)
print("Calificación promedio Tienda 2:", calificacion_promedio_tienda2)
print("Calificación promedio Tienda 3:", calificacion_promedio_tienda3)
print("Calificación promedio Tienda 4:", calificacion_promedio_tienda4)
```

# Informe Final: Recomendación para la Venta de una Tienda de Alura Store

## Introducción
El propósito de este análisis es ayudar al Sr. Juan a decidir cuál de las cuatro tiendas de Alura Store debe vender para iniciar un nuevo emprendimiento. Se analizaron cinco métricas clave: ingresos totales, distribución de categorías de productos, calificaciones promedio de los clientes, productos más y menos vendidos, y costos de envío promedio. Los datos se procesaron con Pandas y se visualizaron con Matplotlib para identificar la tienda menos eficiente, considerando fortalezas y debilidades.

## Desarrollo

### 1. Ingresos Totales
Se calcularon los ingresos totales sumando los valores de la columna `Precio`:
- **Tienda 1**: $1,150,880,400
- **Tienda 2**: $1,116,343,500
- **Tienda 3**: $1,098,019,600
- **Tienda 4**: $1,038,375,700

*Gráfico*: El gráfico de barras "Ingresos Totales por Tienda" muestra que la Tienda 1 tuvo los mayores ingresos ($1,150,880,400), mientras que la Tienda 4 generó los menores ($1,038,375,700), un 9.8% menos que la Tienda 1.

### 2. Categorías de Productos
Se analizó la cantidad de ventas por categoría con `value_counts()`. Resultados destacados:
- **Tienda 1**:
  - Más vendida: Muebles (465 ventas), Electrónicos (448).
  - Menos vendida: Artículos para el hogar (171), Libros (173).
- **Tienda 2**:
  - Más vendida: Muebles (442), Electrónicos (422).
  - Menos vendida: Artículos para el hogar (181), Libros (197).
- **Tienda 3**:
  - Más vendida: Muebles (499), Electrónicos (451).
  - Menos vendida: Instrumentos musicales (177), Artículos para el hogar (177).
- **Tienda 4**:
  - Más vendida: Muebles (480), Electrónicos (451).
  - Menos vendida: Instrumentos musicales (170), Libros (187).

*Gráfico*: El gráfico circular "Distribución de Categorías - Tienda 1" muestra que Muebles y Electrónicos dominan (~30% combinado). Las cuatro tiendas tienen patrones similares, con Muebles y Electrónicos como categorías fuertes, pero la Tienda 4 tiene menos ventas en Electrodomésticos (254 vs. 312 en Tienda 1).

### 3. Calificaciones Promedio
Se calcularon las calificaciones promedio de la columna `Calificación`:
- **Tienda 1**: 3.98
- **Tienda 2**: 4.04
- **Tienda 3**: 4.05
- **Tienda 4**: 4.00

*Gráfico*: El gráfico de barras horizontales "Calificaciones Promedio por Tienda" indica que la Tienda 3 tiene la mayor satisfacción (4.05), mientras que la Tienda 1 tiene la menor (3.98). La Tienda 4 (4.00) está cerca del promedio, pero no destaca.

### 4. Productos Más y Menos Vendidos
Se identificaron los productos con más y menos ventas en cada tienda:
- **Tienda 1**:
  - Más vendido: Microondas (60 ventas).
  - Menos vendido: Celular ABXY (33 ventas).
- **Tienda 2**:
  - Más vendido: Iniciando en programación (65 ventas).
  - Menos vendido: Juego de mesa (32 ventas).
- **Tienda 3**:
  - Más vendido: Kit de bancas (57 ventas).
  - Menos vendido: Bloques de construcción (35 ventas).
- **Tienda 4**:
  - Más vendido: Cama box (62 ventas).
  - Menos vendido: Guitarra eléctrica (33 ventas).

*Gráfico*: El gráfico de barras "Top 5 Productos Más Vendidos - Tienda 1" destaca productos como Microondas. La Tienda 4 tiene un producto líder sólido (Cama box), pero su producto menos vendido (Guitarra eléctrica, 33) está entre los más bajos, similar a otras tiendas.

### 5. Costos de Envío Promedio
Se calcularon los costos de envío promedio de la columna `Costo de envío`:
- **Tienda 1**: $26,018.61
- **Tienda 2**: $25,216.24
- **Tienda 3**: $24,805.68
- **Tienda 4**: $23,459.46

*Gráfico*: El gráfico de dispersión "Costos de Envío Promedio por Tienda" muestra que la Tienda 1 tiene el costo más alto ($26,018.61), mientras que la Tienda 4 tiene el más bajo ($23,459.46), un 10.8% menos que la Tienda 1.

## Conclusión
Con base en los análisis, recomendamos que el Sr. Juan venda la **Tienda 4**. Las razones son:

- **Ingresos bajos**: La Tienda 4 generó $1,038,375,700, el menor ingreso entre todas, 9.8% menos que la Tienda 1 ($1,150,880,400), indicando un rendimiento financiero débil.
- **Calificaciones promedio**: Su calificación de 4.00 está por debajo de las Tiendas 2 (4.04) y 3 (4.05), sugiriendo menor satisfacción del cliente comparada con las líderes.
- **Categorías y productos**: Aunque tiene fortaleza en Muebles (480 ventas) y Electrónicos (451), su desempeño en Electrodomésticos (254 ventas) es el más bajo entre todas las tiendas, y su producto menos vendido (Guitarra eléctrica, 33 ventas) no destaca frente a otras.
- **Costos de envío**: La Tienda 4 tiene el costo de envío promedio más bajo ($23,459.46), lo cual es positivo, pero no compensa las debilidades en ingresos y calificaciones.

Por otro lado, las Tiendas 1, 2 y 3 muestran fortalezas:
- **Tienda 1**: Mayores ingresos ($1,150,880,400), pero calificación más baja (3.98).
- **Tienda 2**: Buena calificación (4.04) y producto estrella (Iniciando en programación, 65 ventas).
- **Tienda 3**: Mayor calificación (4.05) y sólido desempeño en Muebles (499 ventas).

Vender la Tienda 4 permitirá al Sr. Juan enfocarse en las tiendas más rentables y con mejor aceptación, optimizando recursos para su nuevo emprendimiento.
